```python
# load dependencies
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train.shape
```

```python
model = tf.keras.Sequential([
    layers.Input(shape=(28, 28)), # input shape
    layers.Flatten(), # reshapes multi-dimensional tensor into a single dimensional vector
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)
```

```python
num_samples = 5
img_plts = np.random.choice(len(x_test), num_samples)

for i, idx in enumerate(img_plts):
    plt.subplot(1, num_samples, i + 1)  # 1 row, num_samples columns
    plt.imshow(x_test[idx], cmap='gray')  # grayscale
    plt.title(f"{y_test[idx]}")
    plt.axis('off')

plt.show()
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
  x_train,
  y_train_encoded,
  epochs=10,
  batch_size=32,
  validation_split=0.1,
  verbose=1
)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test_encoded)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_acc}")
```

```python
pred_probs = model.predict(x_test)
pred_labels = np.argmax(pred_probs, axis=1)

cm = confusion_matrix(y_test, pred_labels)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=range(10), yticklabels=range(10))
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("MNIST Confusion Matrix")
plt.show()
```
