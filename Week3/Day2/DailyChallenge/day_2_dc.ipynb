```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import rotate

image_path = r"C:\Users\leibn\Downloads\Flower Color Images Dataset\Flower Color Images Dataset\flower_images\flower_images\0103.png"
original_image = Image.open(image_path)

original_np = np.array(original_image)

def rotate_img(image):
    return rotate(image, 30, reshape=False)
rotated = rotate_img(original_np)

plt.imshow(rotated)
plt.show()

```
