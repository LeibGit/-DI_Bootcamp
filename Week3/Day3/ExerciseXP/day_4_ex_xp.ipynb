Exercise 1

Data visualition is important to be able to clearly visualize adn understand relationships between complex data sets. For example, a line graph is usefull for understanding trends in datasets.

```python
days = [1, 2, 3, 4, 5, 6, 7]
temps = [72, 74, 76, 80, 82, 78, 75]

plt.plot(days, temps, color='red', linewidth=2)
plt.xlabel='Day'
plt.ylabel='Temperature'
plt.title='Temp for day of week'
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert Yes/No to 0/1
df = mental_health_data.copy()
df['Panic'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(8, 5))
sns.stripplot(x=df['Age'], y=df['Panic'], jitter=True)
plt.yticks([0, 1], ['No', 'Yes'])
plt.title('Panic Attacks in Relation to Age')
plt.show()
```
