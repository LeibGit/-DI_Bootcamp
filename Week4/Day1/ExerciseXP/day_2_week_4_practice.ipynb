```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
movie_ratings = np.array([
    [5, 1, 4],
    [4, 4, 2],
    [4, 3, 5],
    [1, 1, 5],
    [3, 2, 1]
])

average_rating = np.mean(movie_ratings, axis=1, dtype=int)
viewer_preferance = np.max(movie_ratings, axis=1)


print(f"Average Rating: {average_rating}")
print(f"Viewer preferance: {viewer_preferance}")
```

```python
matrix_B = np.array([[2, 3], [4, 5]])
matrix_C = np.array([[1, 1], [1, 1]])

sum_matrix = np.add(matrix_B, matrix_C)
diff_matrix = np.subtract(matrix_B, matrix_C)

print(f"Sum of Matrices: {sum_matrix}")
print(f"Diff of Matrices: {diff_matrix}")
```

```python
matrix_1 = np.random.rand(3, 3)
matrix_2 = np.random.rand(3, 3)

matrix_sum = np.add(matrix_1, matrix_2)
matrix_sub = np.subtract(matrix_1, matrix_2)
print(f"Sum: {matrix_sum}")
print(f"Sum: {matrix_sub}")
```

```python
# Temperature data matrix (4 cities, 7 days)
temperature_data = np.array([
    [27, 20, 15, 18, 26, 18, 22],
    [24, 18, 20, 17, 19, 22, 21],
    [23, 23, 27, 25, 16, 21, 22],
    [22, 29, 23, 16, 20, 24, 28]
])

average_temp = np.mean(temperature_data, axis=1, dtype=int)
max_temp = np.max(temperature_data, axis=1)
min_temp = np.min(temperature_data, axis=1)

day_one_avg = np.mean(temperature_data[:,0], dtype=int)
day_seven_avg = np.mean(temperature_data[:,6], dtype=int)

difference = day_one_avg, day_seven_avg

print(f"Average Temp Of each city: {average_temp}")
print(f"Max Temp Of each city: {max_temp}")
print(f"Min Temp Of each city: {min_temp}")
print(difference)
```

```python
data = np.arange(20)
sample = np.random.choice(data, size=10, replace=False)

sample_mean = np.mean(sample)
sample_median = np.median(sample)
sample_varience = np.var(sample)
sample_skewness = skew(sample)
sample_kurtosis = kurtosis(sample)

print(sample)
print(sample_mean)
print(sample_median)
print(sample_varience)
print(sample_skewness)
print(sample_kurtosis)
```

```python
dates = pd.date_range(start='2022-01', end='2022-02', freq='D')

temp_data = np.random.uniform(low=10, high=35, size=len(dates))

humidity = np.random.uniform(low=30, high=90, size=len(dates))

data = {
    'Date': dates, 
    'Tempuratue': temp_data, 
    'Humidity %': humidity
}

df = pd.DataFrame(data)
df.head()
```

```python
temp_mean = np.mean(df['Tempuratue'])
temp_median = np.median(df['Tempuratue'])
temp_std = np.std(df['Tempuratue'])

humidity_mean = np.mean(df['Humidity %'])
humidity_median = np.median(df['Humidity %'])
humidity_std = np.std(df['Humidity %'])

print(temp_mean)
print(temp_median)
print(temp_std)

print(humidity_mean)
print(humidity_median)
print(humidity_std)
```

Use a rolling window of 7 days to calculate the moving average for both temperature and humidity.
[bonus] Plot the original data and the 7-day moving average on the same graph for comparison.

```python
df['Temp_MA7'] = df['Tempuratue'].rolling(window=7).mean()
df['Humidity_MA7'] = df['Humidity %'].rolling(window=7).mean()
```
