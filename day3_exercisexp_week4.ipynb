```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_val, p_val = stats.ttest_ind(data1, data2)
print(f"T_val: {t_val}")
print(f"P_val: {p_val}")

print(f"Mean1: {data1.mean()}")
print(f"Mean2: {data2.mean()}")
```

exercise 5

```python
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150_000, 200_000, 210_000, 250_000, 280_000]


slope, intercept, r, p, std_err = stats.linregress(house_sizes, house_prices)

def myfunc(house_sizes):
  return slope * house_sizes + intercept

mymodel = list(map(myfunc, house_sizes))

print(f"relationship in (r): {r}")
print(f"line slope: {slope}")
print(f"intercept: {intercept}")

def get_prediction(n):
  return (n * slope) + intercept


plt.scatter(house_sizes, house_prices)
plt.plot(house_sizes, mymodel)
plt.ylabel("House Prices(In Thousands)")
plt.xlabel("square meters")
plt.title("Price based on square meters")
plt.show()
get_prediction(90)


# the slope represents how much the price increases in relation to the square meter increase
```

exercise 6

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

fertilizer_1_mean = np.mean(fertilizer_1)
fertilizer_2_mean = np.mean(fertilizer_2)
fertilizer_3_mean = np.mean(fertilizer_3)

print(f"Mean 1: {fertilizer_1_mean}")
print(f"Mean 2: {fertilizer_2_mean}")
print(f"Mean 3: {fertilizer_3_mean}")


f_val, p_val = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("f-value", f_val)
print("p-value", p_val)

# the different types of fertilizer are significant
```

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

pearson_corr, pearson_p = stats.pearsonr(data['age'], data['income'])

print(f"Pearson correlation", pearson_corr)
print(f"p-value", pearson_p)

spearman_corr, spearman_p = stats.spearmanr(data['age'], data['income'])

print(f"Spearman correlation", spearman_corr)
print(f"p-value", spearman_p)
```
