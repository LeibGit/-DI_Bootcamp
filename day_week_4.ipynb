```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
total_bills = tips['total_bill'].values
total_bills[:10]
```

```python
total_tips = tips['tip'].values
total_tips[:10]
```

BILL STATS

```python
total_bills.mean() # average bill
```

```python
total_bills.max() # the max
```

```python
total_bills.min() # the min
```

TIP STATS

```python
total_tips.mean() # average tip
```

```python
total_tips.max() # max tips
```

```python
total_tips.min() # min tip
```

```python
high_bill_map = total_bills > 30
total_bills[high_bill_map]
```

```python
small_tips = total_tips < 2
total_tips[small_tips]
```

```python
total_tips[small_tips].mean() # average of the small tips
```

```python
total_tips[small_tips].size # amount of small tips
```
