```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(y_train[:5])
print(y_test[:5])
```

```python
y_train_oh = to_categorical(y_train)
y_test_oh = to_categorical(y_test)
```

```python
model_full = Sequential([
    Flatten(input_shape=(28, 28)), # Added Flatten layer
    Dense(28, activation="relu"),
    Dense(64, activation="relu"),
    Dense(10, activation="softmax")
])

model_full.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model_full.fit(
    x=x_train,
    y=y_train_oh,
    epochs=3,
    batch_size=32,
    validation_split=0.2,
    )

test_loss_full, test_accuracy_full = model_full.evaluate(x=x_test, y=y_test_oh)
print(f"Test Accuracy: {test_accuracy_full}")
print(f"Test Loss: {test_loss_full}")
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(
    x=x_train,
    y=y_train_oh,
    epochs=3,
    batch_size=32,
    validation_split=0.2,
    )

test_loss, test_accuracy = model.evaluate(x=x_test, y=y_test_oh)
print(f"Test Accuracy: {test_accuracy}")
print(f"Test Loss: {test_loss}")
```

```python
if test_accuracy > test_accuracy_full:
  print("The CNN model performed better.")
else:
  print("The full model trained better.")
```
