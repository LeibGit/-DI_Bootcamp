exercise 6

```python
# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

print(productivity_before)
print(productivity_after)

average_diff = np.median(productivity_after) - np.median(productivity_before)
before_max = np.max(productivity_before)
after_max = np.max(productivity_after)

before_min = np.min(productivity_before)
after_min = np.min(productivity_after)


print(average_diff)

print(before_max)
print(after_max)

print(before_min)
print(after_min)

# As seen by the data there is a clear improvevemeny in productivity given training
```

```python
arr_1 = np.array([1, 45, 300, 6, 70])
arr_2 = np.array([12, 20, 3, 667, 7])

greater_than = {}
less_than = {}

for i, num in enumerate(arr_1):
  if num > arr_2[i]:
    greater_than[i] = num
  elif num < arr_2[i]:
    less_than[i] = num
  else:
    print("equal")
print(greater_than)
print(less_than)
```
