```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
df = pd.DataFrame(data)
df.head()
```

```python
df.dtypes
```

Each number here is classified as a qunatitative or qualitative type based on if it's type returned either or a float or an int. If it did, I counted it as a quantitative type, otherwise I resorted to classyfing it as a qualitative type.

```python
qualitative_type = 0
quantitative_type = 0

for t in df.dtypes:
  if t == np.int64 or t == np.float64:
    quantitative_type += 1
  else:
    qualitative_type += 1
print(quantitative_type)
print(qualitative_type)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("sleep.csv")
df = pd.DataFrame(data)
df.head(n=10)
```

I chose to analyze the trend of sleep overtime, to see if average sleep has increased or decreased for individuals overtime.

```python
df_yearly = df.groupby("Year")["Avg hrs per day sleeping"].mean()
plt.plot(df_yearly.index, df_yearly.values)
plt.xlabel("Year")
plt.ylabel("Avg hrs per day sleeping")
plt.title("Trend of sleep overtime.")
plt.show()
```
