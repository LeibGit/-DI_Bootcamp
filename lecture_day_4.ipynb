```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, color='green', linestyle='dashed', linewidth=2, marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```

```python
catagories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [20, 35, 30, 15]

plt.bar(catagories, values, color=['red', 'blue', 'purple', 'orange'], edgecolor='black', hatch='/')
plt.title('Fruit Sales Data')
plt.xlabel('Fruit')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np

data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
brands = ["Apple", "Samsung", "Huawei", "Xiaomi", "Others"]
market_share = [40, 30, 15, 10, 5]

plt.pie(market_share, labels=brands, autopct='%1.1f%%')
plt.title("Smartphone market share 2023")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title('First Subplot')
ax2.plot([1, 2, 3], [1, 2, 3])
ax2.set_title('Second Subplot')
plt.show()
```

```python
x = [2, 4, 6, 8, 10]
y = []
for i in x:
  y.append(i**2)

plt.plot(x, y, color='purple', linewidth=4)
plt.title("Power of X")
plt.xlabel='x-value'
plt.ylabel = 'x-squared'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], shade=True, color='purple')
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=data, kind="scatter", color="green")
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Survival rate by class')
plt.show()
```
