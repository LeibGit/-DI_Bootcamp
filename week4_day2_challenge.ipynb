```python
import pandas as pd
import numpy as np
from scipy.stats import *
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("global_data.csv")
df.head()
```

```python
df.columns
```

```python
df.isnull()
```

```python
# changin non present values to NA
df = df.fillna("N/A")
```

```python
generation_cols = [
    'estimated_generation_gwh_2013',
    'estimated_generation_gwh_2014',
    'estimated_generation_gwh_2015',
    'estimated_generation_gwh_2016',
    'estimated_generation_gwh_2017'
]

for col in generation_cols:
    # Convert 'N/A' strings to NaN and ensure the column is numeric
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # Fill NaN values with the median of the column
    df[col] = df[col].fillna(df[col].median())
```

```python
df.head()
```

```python
for col in df.columns:
  if df[col].dtype == int or df[col].dtype == float:
    print(f"{col} Average: {df[col].median()}")
    print(f"{col} Mean: {df[col].mean()}")
    print(f"{col} Mean: {df[col].mean()}")
    print(f"{col} Standard deviation: {df[col].std()}")
  else:
    continue
```

```python
df['country_long'].value_counts()
df['primary_fuel'].value_counts()


grouped = df.groupby(['country_long', 'primary_fuel']).size().nlargest().unstack()


grouped.plot(kind="bar", figsize=(5, 6))
plt.show()
```

```python
fuel_capacity_rel = df.groupby("primary_fuel")["capacity_mw"].nlargest()

fuel_capacity_rel.plot()
plt.show()
```

```python
fuel_types = df.groupby("primary_fuel")['capacity_mw'].mean()
print(fuel_types)
```

Nuclear energy seems to reflect the highest effeciency. Power outpuy differs significantly

```python
sns.barplot(fuel_types.nlargest())
```

```python
trend = df.groupby(['commissioning_year', 'primary_fuel']).size().unstack()

trend.plot(figsize=(10,6))
plt.title("Commissioning Trend by Fuel Type")
plt.xlabel("Year")
plt.ylabel("Number of Plants")
plt.legend(title='Primary Fuel')
plt.show()
```
